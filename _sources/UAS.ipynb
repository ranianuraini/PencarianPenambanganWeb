a. Crawling berita online berdasarkan kategori

```python
import requests
from bs4 import BeautifulSoup

def crawl_detik_news(category_url):
    # Mengunduh halaman kategori
    response = requests.get(category_url)
    if response.status_code != 200:
        print(f"Gagal mengunduh halaman {category_url}")
        return []

    # Membuat objek BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari tautan berita
    news_links = []
    for article in soup.find_all('article'):
        link = article.find('a')['href']
        news_links.append(link)

    return news_links

# Kategori yang akan di-crawl
categories = ['https://sport.detik.com/', 'https://travel.detik.com/', 'https://food.detik.com/']

# Crawl berita untuk setiap kategori
for category_url in categories:
    news_links = crawl_detik_news(category_url)

    # Menampilkan hasil
    print(f"\nBerita dari kategori {category_url}:")
    for link in news_links:
        print(link)
```

```python
import csv

# Menyiapkan file CSV
csv_file_path = 'detik_news.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Nomor', 'Kategori', 'Judul Berita', 'Link'])

    # Crawl berita untuk setiap kategori
    for category_url in categories:
        news_links = crawl_detik_news(category_url)

        # Menyimpan hasil ke dalam file CSV
        for i, link in enumerate(news_links, start=1):
            csv_writer.writerow([i, category_url, link.split("/")[-1], link])

# Menampilkan pesan jika penyimpanan berhasil
print(f"Data telah disimpan dalam file CSV: {csv_file_path}")
```
