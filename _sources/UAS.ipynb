```python
import requests
from bs4 import BeautifulSoup
import csv

def crawl_detik_news(category_url):
    # Mengunduh halaman kategori
    response = requests.get(category_url)
    if response.status_code != 200:
        print(f"Gagal mengunduh halaman {category_url}")
        return []

    # Membuat objek BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari tautan berita
    news_links = []
    for article in soup.find_all('article'):
        link = article.find('a')['href']
        news_links.append(link)

    return news_links

# Mapping URL kategori ke nama kategori
category_mapping = {
    'https://sport.detik.com/': 'olahraga',
    'https://travel.detik.com/': 'travel',
    'https://food.detik.com/': 'makanan'
}

# Kategori yang akan di-crawl
categories = list(category_mapping.keys())

# Menyiapkan file CSV
csv_file_path = 'detik_newss.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Nomor', 'Kategori', 'Judul Berita', 'Link'])

    # Crawl berita untuk setiap kategori
    for category_url in categories:
        news_links = crawl_detik_news(category_url)

        # Menyimpan hasil ke dalam file CSV
        category_name = category_mapping[category_url]
        for i, link in enumerate(news_links, start=1):
            csv_writer.writerow([i, category_name, link.split("/")[-1], link])

# Menampilkan pesan jika penyimpanan berhasil
print(f"Data telah disimpan dalam file CSV: {csv_file_path}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def crawl_detik_news(category_url, category_name, total_data):
    base_url = 'https://www.detik.com/'

    news_data = []

    for page in range(1, total_data // 10 + 1):
        url = f'{category_url}?page={page}'
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            articles = soup.find_all('article')

            for i, article in enumerate(articles):
                title = article.find('h2').get_text(strip=True)
                link = base_url + article.find('a')['href']

                # Get content of the news
                content_response = requests.get(link)
                if content_response.status_code == 200:
                    content_soup = BeautifulSoup(content_response.text, 'html.parser')
                    content_paragraphs = content_soup.find_all('p')
                    content = ' '.join([p.get_text(strip=True) for p in content_paragraphs])
                else:
                    content = ''

                news_data.append({
                    'number': i + 1 + (page - 1) * 10,
                    'category': category_name,
                    'title': title,
                    'link': link,
                    'content': content
                })

    return news_data

```

```python
def save_to_csv(data, filename='detik_news_with_content.csv'):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['number', 'category', 'title', 'link', 'content']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for row in data:
            writer.writerow(row)

if __name__ == '__main__':
    total_data_per_category = 1000
    sport_news = crawl_detik_news('https://sport.detik.com/', 'olahraga', total_data_per_category)
    travel_news = crawl_detik_news('https://travel.detik.com/', 'travel', total_data_per_category)
    food_news = crawl_detik_news('https://food.detik.com/', 'makanan', total_data_per_category)

    all_news = sport_news + travel_news + food_news
    save_to_csv(all_news)
```
