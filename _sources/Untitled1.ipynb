# Implementasi Reduksi Dimensi PTA Dataset

```python
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Kuliah /Semester 7/Pencarian Dan Penambangan Web/ppw/Dataset/crawling_pta_labeled.csv')
```

```python
data = data.dropna()
```

```python
df = pd.read_csv('/content/drive/MyDrive/Kuliah /Semester 7/Pencarian Dan Penambangan Web/ppw/Term Frequensi.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
X = df.drop(['Dokumen'], axis=1)
X
```

```python
k = 3
alpha = 0.1
beta = 0.2

lda = LatentDirichletAllocation(n_components=k, doc_topic_prior=alpha, topic_word_prior=beta)
proporsi_topik_dokumen = lda.fit_transform(X)
```

```python
dokumen = df['Dokumen']
output_proporsi_TD = pd.DataFrame(proporsi_topik_dokumen, columns=['Topik 1', 'Topik 2', 'Topik 3'])
output_proporsi_TD.insert(0,'Dokumen', dokumen)
output_proporsi_TD.insert(len(output_proporsi_TD.columns),'Label', data['Label'])
output_proporsi_TD
```

```python
output_proporsi_TD = output_proporsi_TD.isnull().sum()
```

```python
# Output distribusi kata pada topik
distribusi_kata_topik = pd.DataFrame(lda.components_)
distribusi_kata_topik
```

# Modeling Naive Bayes (Term Frequensi)

```python
# Impor library yang diperlukan
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Memuat dataset Iris
X = proporsi_topik_dokumen
y = output_proporsi_TD['Label']

# Memisahkan dataset menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Membuat model Naive Bayes
model = GaussianNB()

# Melatih model pada data pelatihan
model.fit(X_train, y_train)

# Melakukan prediksi pada data pengujian
y_pred = model.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)

# Menampilkan laporan klasifikasi
print("Laporan Klasifikasi:")
print(classification_report(y_test, y_pred))

# Menampilkan matriks kebingungan
confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matriks:")
print(confusion)
```
